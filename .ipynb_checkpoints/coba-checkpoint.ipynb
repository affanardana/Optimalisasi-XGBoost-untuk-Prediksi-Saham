```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import TimeSeriesSplit
import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
tes = pd.read_csv('./AMZN.csv')
saham = pd.read_csv('./AMZN.csv')
saham.head()
```

```python
saham.info()
```

```python
saham['Date'] = pd.to_datetime(saham['Date'], format='%d-%m-%Y')
tes['Date'] = pd.to_datetime(tes['Date'], format='%d-%m-%Y')
saham.head()
```

```python
plt.figure(figsize=(15,5))
plt.plot(saham['Date'], saham['Adjusted Close'])
```

```python
saham.info()
```

```python
tes.info()
```

```python
saham = saham.loc[(saham['Date'] > '1998-01-01') & (saham['Date'] < '2021-12-31')]
tes = saham.loc[(saham['Date'] > '1998-01-01') & (saham['Date'] < '2021-12-31')]
```

```python
saham['Date'].count()
```

```python
tes['Date'].count()
```

```python
cols = ["Low", "Open", "Volume", "High", "Close", "Adjusted Close"]

fig, axs  = plt.subplots(3, 2, figsize=(15, 15))

for col, ax in zip(cols, axs.flat):
        ax.plot(saham["Date"], saham[col])

        ax.set_xlabel("Date")
        ax.set_ylabel(col)

plt.show()
```

```python
print(f"Tanggal data pertama: {saham['Date'].min()}")
print(f"Tanggal data terakhir: {saham['Date'].max()}")
```

```python
print(f"Tanggal data pertama: {tes['Date'].min()}")
print(f"Tanggal data terakhir: {tes['Date'].max()}")
```

```python
saham.describe()
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))

saham = scaler.fit_transform(saham.iloc[:, 1:])
```

```python
len(saham)
```

```python
tes['Date'].count()
```

```python
saham = pd.DataFrame(saham, columns=['Low', 'Open', 'Volume', 'High', 'Close', 'Adjusted Close'])
saham.head()
```

```python
saham = pd.DataFrame(saham, columns=['Low', 'Open', 'Volume', 'High', 'Close', 'Adjusted Close'])
saham.insert(0, 'Date', tes['Date'], True)
saham.head()
```

```python
saham.describe()
```

```python
saham.info()
```

```python
train = saham.loc[saham.index < 474]
test = saham.loc[saham.index >= 474]

#tgl_test = saham.iloc[:,0].loc[saham['Date'] >= '2019-01-01']

X_train = train.iloc[:, [0,1,3,4]]
y_train = train.iloc[:,5]

X_test = test.iloc[:, [0,1,3,4]]
y_test = test.iloc[:,5]
```

```python
train = saham.loc[saham['Date'] < '2022-06-16'] # 2022-06-16
test = saham.loc[saham['Date'] >= '2022-06-16']

tgl_test = saham.iloc[:,0].loc[saham['Date'] >= '2022-06-16']

X_train = train.iloc[:, [1,2,4,5]]
y_train = train.iloc[:,6]

X_test = test.iloc[:, [1,2,4,5]]
y_test = test.iloc[:,6]
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
print()
print(tgl_test.shape)
```

```python
model = xgb.XGBRegressor(n_estimators=500, learning_rate=0.01, max_depth=3)

model.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)], verbose=50)

y_pred = model.predict(X_test)
```

```python
df_pred = pd.DataFrame({'pred': y_pred, 'real': y_test})
df_pred = df_pred.reset_index()
df_pred.head()
```

```python
df_pred = pd.DataFrame({'date': tgl_test, 'pred': y_pred, 'real': y_test})
df_pred = df_pred.reset_index()
df_pred.head()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_pred.index, df_pred['real'], 'o', color='green')
plt.plot(df_pred.index, df_pred['real'], color='green', linestyle='-')
plt.plot(df_pred.index, df_pred['pred'], '^', color='red')
plt.plot(df_pred.index, df_pred['pred'], color='red', linestyle='-')
```
